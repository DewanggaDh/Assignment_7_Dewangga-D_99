```python
df_spark = spark.createDataFrame(df)
```

```python
df_spark.show()
```

```python
df_1 = df[['Customer','SalesTransactionID','SalesItem']]
```

```python
df_1
```

```python
df_2 = df_1.groupby('SalesTransactionID')['SalesItem'].apply(lambda x: x.reset_index (drop = True)).unstack().reset_index()
df_2.drop('SalesTransactionID',axis=1, inplace=True)
df_2.head()
```

```python
items = df_2[0].unique()
```

```python
encoded_vals = []
def custom():
    for index, row in df_2.iterrows():
        labels = {}
        uncommons = list(set(items) - set(row))
        commons = list(set(items).intersection(row))
        for uc in uncommons:
            labels[uc] = 0
        for com in commons:
            labels[com] = 1
        encoded_vals.append(labels)
custom()
ohe_df = pd.DataFrame(encoded_vals)
```

```python
ohe_df
```

```python
freq_items = apriori(ohe_df, min_support=0.006, use_colnames=True)
freq_items
```

```python
from pyspark.sql import functions as F
from pyspark.ml.fpm import FPGrowth
```

```python
basketdata = df_spark.dropDuplicates(['SalesTransactionID', 'SalesItem']).sort('SalesTransactionID')
basketdata = basketdata.groupBy("SalesTransactionID").agg(F.collect_list("SalesItem")).sort('SalesTransactionID')
```

```python

fpGrowth = FPGrowth(itemsCol="collect_list(SalesItem)", minSupport=0.006, minConfidence=0.006) 
model = fpGrowth.fit(basketdata)

model.freqItemsets.show()
items = model.freqItemsets

model.associationRules.show()
rules = model.associationRules

model.transform(basketdata).show()
transformed = model.transform(basketdata)
```

```python
result_pdf = items.select("*").toPandas()
result_pdf.head()
```

```python

rules_pdf = rules.select("*").toPandas()
rules_pdf.head()
```

```python
transformed_pdf = transformed.select("*").toPandas()
transformed_pdf.head()
```
