```python
fruits = [['apple', 'orange', 'banana', 'pineapple', 'grapes'],
          ['orange', 'grapes', 'apple'],
          ['banana'],
          ['grapes', 'pineapple', 'kiwi', 'apple'],
          ['kiwi', 'orange', 'apple'],
          ['orange', 'banana'],
          ['apple', 'orange', 'kiwi', 'watermelon', 'lemon'],
          ['lemon', 'melon', 'banana', 'grapes']]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(fruits).transform(fruits)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df)
```

```python
apriori(df, use_colnames=True)
```

```python
frequent_itemsets = apriori(df, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 1) &
                   (frequent_itemsets['support'] >= 0.6) ]
```

```python
frequent_itemsets[ frequent_itemsets['itemsets'] == {'apple'} ]
```

```python
oht_ary = te.fit(fruits).transform(fruits, sparse=True)
sparse_df = pd.DataFrame.sparse.from_spmatrix(oht_ary, columns=te.columns_)
sparse_df
```

```python
apriori(sparse_df, min_support=0.6, use_colnames=True)
```
